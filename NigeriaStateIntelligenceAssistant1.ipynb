### Create RAG Documents

RAG systems work better with natural language documents, although the source data is structured JSON, we will transform it into descriptive text. This process is called: Structured-to-unstructured transformation.

```python
# The API does not accept some state names directly from /fetch. Results to manual

states = [
    "Abia",
    "Adamawa",
    "Akwa Ibom",
    "Anambra",
    "Bauchi",
    "Bayelsa",
    "Benue",
    "Borno",
    "Cross River",
    "Delta",
    "Ebonyi",
    "Edo",
    "Ekiti",
    "Enugu",
    "FCT",
    "Gombe",
    "Imo",
    "Jigawa",
    "Kaduna",
    "Kano",
    "Katsina",
    "Kebbi",
    "Kogi",
    "Kwara",
    "Lagos",
    "Nasarawa",
    "Niger",
    "Ogun",
    "Ondo",
    "Osun",
    "Oyo",
    "Plateau",
    "Rivers",
    "Sokoto",
    "Taraba",
    "Yobe",
    "Zamfara"
]

documents = []

failed_states = []

for state in states:

    url = (
        f"https://nga-states-lga.onrender.com/?state={state}"
    )

    response = requests.get(url)

    try:

        lgas = response.json()

        if isinstance(lgas, list):

            document = f"""
            State: {state}

            Local Government Areas:
            {', '.join(lgas)}
            """

            documents.append(document)

        else:

            failed_states.append(state)

    except Exception:

        failed_states.append(state)

print(f"Documents created: {len(documents)}")

print(f"Failed states: {failed_states}")

print(documents[0])
```

```python
# API Verification Framework

import requests
from pprint import pprint

def inspect_api(name, url):
    
    try:
        response = requests.get(url, timeout=20)
        print(f" status code: {response.status_code}")
        
        data = response.json()
        
        print(f"data type: {type(data)}")
        
        
        # List Response
        if isinstance(data, list):
            
            print(f"List length: {len(data)}")
            
            pprint(data[0])
            
            
        # Dictionary Response
        elif isinstance(data, dict):
            pprint(list(data.keys())[:20])
            
            pprint(data)
            
        else:
            print(data)
            
    except Exception as e:
        
        print(f"FAILED: {e}")
```

### Create Independent Ingestion Modules

```python
# State module
def ingest_states():

    documents = []

    try:

        states = [
            "Lagos",
            "Kano",
            "Kaduna"
        ]

        for state in states:

            url = (
                f"https://nga-states-lga.onrender.com/?state={state}"
            )

            response = requests.get(url)

            lgas = response.json()

            document = {
                "source": "states_api",
                "entity_type": "state",
                "entity_name": state,
                "content": f"""
                State: {state}

                LGAs:
                {', '.join(lgas)}
                """
            }

            documents.append(document)

    except Exception as e:

        print(f"States ingestion failed: {e}")

    return documents
```

```python
# Universities
def ingest_universities():

    documents = []

    try:

        url = (
            "http://universities.hipolabs.com/search?country=Nigeria"
        )

        data = requests.get(url).json()

        for uni in data:

            document = {
                "source": "universities_api",
                "entity_type": "university",
                "entity_name": uni.get("name"),
                "content": f"""
                University:
                {uni.get('name')}

                Domain:
                {uni.get('domains')}

                Website:
                {uni.get('web_pages')}
                """
            }

            documents.append(document)

    except Exception as e:

        print(f"University ingestion failed: {e}")

    return documents
```

### Master Ingestion Pipeline

```python
all_documents = []

all_documents.extend(
    ingest_states()
)

all_documents.extend(
    ingest_universities()
)

print(f"Total documents: {len(all_documents)}")

print(all_documents[15])
```

### Create Text Corpus
> Extract raw texts for embeddings

```python
corpus = [
    doc["content"]
    for doc in all_documents
]

print(f"Corpus Size: {len(corpus)}")
```

```python
print(corpus[0][:500])
```

### Load Embedding Model

```python
embedding_model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)
print("model loaded")
```

### Generate Embeddings

```python
embeddings = embedding_model.encode(
    corpus,
    show_progress_bar=True
)

print(embeddings.shape) # MiniLM generates 384 dimensional vectors
```

### Create FAISS Index

```python
embedding_dimensiion = embeddings.shape[1]

index = faiss.IndexFlatL2(
    embedding_dimensiion
)

index.add(
    np.array(embeddings).astype("float32")
)

print(f"FAISS Vectors Stored: {index.ntotal}")
```

### Save Vector Index

```python
faiss.write_index(
    index,
    "NSIA.index"
)
print("saved model")
```

### Create BM25 Retriever

```python
tokenized_corpus = [
    doc.split()
    for doc in corpus
]

bm25 = BM25Okapi(
    tokenized_corpus
)

print("BM25 initialized")
```

### Hybrid Retriever

```python
def hybrid_retrieve(query, top_k=5):

    # Dense retrieval
    query_embedding = embedding_model.encode([query])

    distances, indices = index.search(
        np.array(query_embedding).astype("float32"),
        top_k
    )

    dense_results = [
        all_documents[idx]
        for idx in indices[0]
    ]

    # BM25 retrieval
    tokenized_query = query.split()

    bm25_scores = bm25.get_scores(
        tokenized_query
    )

    bm25_indices = np.argsort(
        bm25_scores
    )[-top_k:]

    keyword_results = [
        all_documents[idx]
        for idx in bm25_indices
    ]

    # Merge results
    combined = dense_results + keyword_results

    # Remove duplicates
    unique_results = []

    seen = set()

    for item in combined:

        content = item["content"]

        if content not in seen:

            unique_results.append(item)

            seen.add(content)

    return unique_results[:top_k]
```

```python
def build_prompt(query, retrieved_docs):
    """
    Constructs a structurally sound prompt for Mistral 7B.
    Wraps context and user query inside structural [INST] tags to enforce 
    strict adherence to retrieved data.
    """
    # Consolidate text content from retrieved document dict arrays
    context = "\n\n".join([
        doc["content"]
        for doc in retrieved_docs
    ])

    # Mistral requires instructions inside [INST] tags. 
    # The trailing 'Answer:' goes outside to trigger generation text naturally.
    prompt = f"""[INST]
You are a Nigeria Intelligence Assistant.

Answer the user's Question using ONLY the factual statements listed under the Retrieved Information block below. 
Do not assume or bring in external knowledge.

If the exact answer is not explicitly mentioned in the text below, respond with exactly:
"I could not find that information."

Retrieved Information:
{context}

Question:
{query}
[/INST]
Answer:"""

    return prompt
```
